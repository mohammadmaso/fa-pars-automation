```python
import pandas as pd
```

```python
data = pd.read_csv('fa-pars.csv')
data.head()
```

```python
data = data.drop(axis=1, index=[0,1])
data = data[data.columns[:-1]]
data.columns = ["LAST_NAME", "ADDRESS", "PHONE", "MOBILE_PHONE", "MODEL", "ID", "DESCRIPTION", "TECHNICIAN"]

data.head()



```

```python
data = data.fillna(-1)
data = data[data.columns[1:]]

data.head()
```

```python
data.index = data.ID.reset_index(drop=True)
```

```python
data.to_csv('cleared_data.csv')
```
